```python

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
def load_data(file_path):
    return pd.read_csv(file_path)

df = load_data(file_path="../Data/raw/loan_approval_dataset.csv")
df.head()
```

```python
# Create a Working Copy

df_clean = df.copy()
df_clean.head()
```

```python
# Standardize Column Names

print(df_clean.columns.tolist())

# Standardize column names

df_clean.columns = (df_clean.columns.str.strip().str.lower().str.replace(" ", "_"))
print(df_clean.columns.to_list())
```

```python
# Check Missing Values

df_clean.isnull().sum()

# Better Business View

missing_values = pd.DataFrame({
    "Missing Values": df_clean.isnull().sum(),
    "Percentage": (df_clean.isnull().sum() / len(df_clean)) * 100
})

missing_values

# Sort by Missing Values

missing_values = missing_values.sort_values(
    by="Missing Values",
    ascending=False
)

missing_values
```

```python
# Duplicate Record Analysis

duplicates = df_clean[df_clean.duplicated()]
duplicates

# Duplicates Found
# df_clean.drop_duplicates(inplace=True)
```

```python
# Check for Extra Spaces in Categorical Values

categorical_columns = df_clean.select_dtypes(include="object").columns
print(categorical_columns)

for col in categorical_columns:
    print("=" * 50)
    print(col)
    print(df_clean[col].unique())
```

```python
# If Extra Spaces Exist

for col in categorical_columns:
    df_clean[col] = df_clean[col].str.strip()

for col in categorical_columns:
    print("=" * 50)
    print(col)
    print(df_clean[col].unique())
```

```python
# Check Constant Columns

constant_columns = []

for col in df_clean.columns:
    if df_clean[col].nunique() == 1:
        constant_columns.append(col)

print(constant_columns)


# If Constant Columns Exist
# df_clean.drop(columns=constant_columns, inplace=True)
```
